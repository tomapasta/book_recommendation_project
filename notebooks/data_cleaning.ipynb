```python
import pandas as pd
```

```python
df_ws= pd.read_csv('../data/processed/goodreads_books.csv')
df_bio=pd.read_csv('../data/processed/Open_Library_Biography_400.csv')
df_fic= pd.read_csv('../data/processed/Open_Library_Fiction_1000.csv')
df_his = pd.read_csv('../data/processed/Open_Library_History_600.csv')
```

```python
df_ws.info()
```

```python
df_api = pd.concat([df_bio, df_fic, df_his], ignore_index=True)
```

```python
df_api.info()
```

```python
df_api_renamed = df_api.rename(columns={
    "Title": "title",
    "Author": "author",
    "Genre": "genres",
    "Rating_Average": "rating",
    "Rating_Counts": "rating_count",
    "Description": "description",
    "Page_Numbers": "pages",
    "ISBN": "isbn",
    "Language": "language",
    "Published_Year": "published_year",
    "Cover_URL": "image",
    "Book_Link": "book_url"
})
```

```python
common_cols = [
    "title", "author", "rating", "rating_count",
    "description", "isbn", "image",
    "book_url", "pages", "published_year",
    "language", "genres"
]

df_ws_clean = df_ws[common_cols]
df_api_clean = df_api_renamed[common_cols]
```

```python
df_all = pd.concat(
    [df_ws_clean, df_api_clean],
    ignore_index=True
)
```

```python
df_all.info()
df_all.head()
```

```python
df_ws_clean["source"] = "web_scraping"
df_api_clean["source"] = "api"

df_all = pd.concat(
    [df_ws_clean, df_api_clean],
    ignore_index=True
)
```

```python
df_all["title_norm"] = (
    df_all["title"]
    .str.lower()
    .str.strip()
)
```

```python
df_all["author_norm"] = (
    df_all["author"]
    .str.lower()
    .str.strip()
)
```

```python
df_all_dedup = df_all.drop_duplicates(
    subset=["title_norm", "author_norm"],
    keep="first"
)
```

```python
df_all_dedup = df_all_dedup.drop(
    columns=["title_norm", "author_norm"]
)
```
